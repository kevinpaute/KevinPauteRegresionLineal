## Preprocesamiento del dataset. Transformaciones previas necesarias para la modelación

```python
## Se define una lista con las variables de entrada del dataset 
var_entradas = ['X1 transaction date','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude'] 
# Se asigna a la variable 'X' las variables de entrada
X = df_rs[var_entradas] 
# Se asigna a la variable 'y' la variable objetivo
y = df_rs[['Y house price of unit area']] 

'''
    Estandarización de la data
'''
# Se importa la libreria StandardScaler para estandarizar la data
from sklearn.preprocessing import StandardScaler 
# Crear un objeto de la clase StandardScaler para estandarizar las variables de entrada
VariablesEntradaScaler=StandardScaler() 
# Crear un objeto de la clase StandardScaler para la variable objetivo
VariableObjetivoScaler=StandardScaler() 

'''
    Almacenamiento del objeto de ajuste para referencia posterior
'''

# Se entrena el objeto VariablesEntradaScaler
VariablesEntradaFit=VariablesEntradaScaler.fit(X) 
# Se entrena el objeto VariableObjetivoScaler
VariableObjetivoFit=VariableObjetivoScaler.fit(y) 

'''
    Generación de los valores estandarizados de X e y
'''

# Se aplica el transformador de 'X' para estandarizar las variables de entrada
X=VariablesEntradaFit.transform(X)
# Se aplica el transformador de 'y' para estandarizar la variable objetivo 
y=VariableObjetivoFit.transform(y) 
```

## División del dataset en datos de entrenamiento y datos de test 

```python
#Código que realice la división en entrenamiento y test, de acuerdo con la estretgia de evaluación planeada. Describa cuál es.
#Importar libre train_test_split
from sklearn.model_selection import train_test_split

# 0.3 = 30% de los datos para el test
X_train, X_test, y_train, y_test = train_test_split(df_rs.drop('Y house price of unit area', axis=1), y, test_size=0.3, random_state=1)  
#Muestra el tamaño de los datos de entrenamiento y test
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Regresión Lineal

```python
'''
    Regresiòn lineal
'''
fig, (ax1,ax2) = plt.subplots(1, 2,figsize=(14,7)) # Se crea una figura con 2 subplots de 14x7
fig, (ax3,ax4) = plt.subplots(1, 2,figsize=(14,7)) # Se crea una figura con 2 subplots de 14x7
fig, (ax5,ax6) = plt.subplots(1, 2,figsize=(14,7)) # Se crea una figura con 2 subplots de 14x7

# Se grafica la relación entre las variables de entrada y la variable de salida
sns.regplot(X_train['X1 transaction date'], y_train, ax=ax1)  
sns.regplot(X_train['X2 house age'], y_train, ax=ax2) 
sns.regplot(X_train['X3 distance to the nearest MRT station'], y_train, ax=ax3)
sns.regplot(X_train['X4 number of convenience stores'], y_train, ax=ax4)
sns.regplot(X_train['X5 latitude'], y_train, ax=ax5)
sns.regplot(X_train['X6 longitude'], y_train, ax=ax6)

# Se muestra la figura
plt.show()
```
